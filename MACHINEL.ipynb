```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
###REGRESSIONE LINEARE
# Carica il dataset
df = pd.read_csv("C:/Users/Utente/Desktop/cakio/dcereijo-player-scores/data/transfers.csv")

# Rimuovi eventuali righe con valori mancanti
df = df.dropna()

# Seleziona le colonne che verranno utilizzate come features (variabili indipendenti)
features = df[['club_name', 'player_name', 'age', 'position', 'club_involved_name', 'transfer_period', 'league_name', 'season', 'OverallSeasonClub', 'OverallSeasonClub2']]

# Codifica le variabili categoriche
features_encoded = pd.get_dummies(features)

# Seleziona la variabile target (variabile dipendente)
target = df['fee_cleaned']

# Suddividi il dataset in set di addestramento e di test
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Crea il modello di regressione lineare
model = LinearRegression()

# Addestra il modello
model.fit(X_train, y_train)

# Effettua le previsioni sul set di test
y_pred = model.predict(X_test)

# Crea un DataFrame con i risultati delle previsioni
results_df = pd.DataFrame({'fee_cleaned_predicted': y_pred, 'original_row': X_test.index})

print(results_df)

# Valuta le prestazioni del modello
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
###REGRESSIONE POLINOMIALE
# Carica il dataset
df = pd.read_csv("C:/Users/Utente/Desktop/cakio/dcereijo-player-scores/data/transfers.csv")

# Rimuovi eventuali righe con valori mancanti
df = df.dropna()

# Seleziona le colonne che verranno utilizzate come features (variabili indipendenti)
features =  df[['club_name', 'player_name', 'age', 'position', 'club_involved_name', 'transfer_period', 'league_name', 'season', 'OverallSeasonClub', 'OverallSeasonClub2']]

# Codifica le variabili categoriche
features_encoded = pd.get_dummies(features)

# Seleziona la variabile target (variabile dipendente)
target = df['fee_cleaned']

# Suddividi il dataset in set di addestramento e di test
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Crea le features polinomiali
degree = 2# Grado del polinomio
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

# Crea il modello di regressione lineare
model = LinearRegression()

# Addestra il modello sulla trasformazione polinomiale delle features
model.fit(X_train_poly, y_train)

# Effettua le previsioni sul set di test
y_pred = model.predict(X_test_poly)

# Valuta le prestazioni del modello
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
###RANDOMFOREST/GRADIENTBOOSTING
# Carica il dataset
df = pd.read_csv("C:/Users/Utente/Desktop/cakio/dcereijo-player-scores/data/transfers.csv")

# Rimuovi eventuali righe con valori mancanti
df = df.dropna()

# Seleziona le colonne che verranno utilizzate come features (variabili indipendenti)
features =  df[['club_name', 'player_name', 'age', 'position', 'club_involved_name', 'transfer_period', 'league_name', 'season', 'OverallSeasonClub', 'OverallSeasonClub2']]

# Codifica le variabili categoriche
features_encoded = pd.get_dummies(features)

# Seleziona la variabile target (variabile dipendente)
target = df['fee_cleaned']

# Suddividi il dataset in set di addestramento e di test
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Crea il modello di Random Forest
rf_model = RandomForestRegressor()

# Addestra il modello di Random Forest
rf_model.fit(X_train, y_train)

# Effettua le previsioni con il modello di Random Forest
rf_y_pred = rf_model.predict(X_test)

# Valuta le prestazioni del modello di Random Forest
rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_r2 = r2_score(y_test, rf_y_pred)

print("Random Forest - Mean Squared Error:", rf_mse)
print("Random Forest - R-squared:", rf_r2)

# Crea il modello di Gradient Boosting
gb_model = GradientBoostingRegressor()

# Addestra il modello di Gradient Boosting
gb_model.fit(X_train, y_train)

# Effettua le previsioni con il modello di Gradient Boosting
gb_y_pred = gb_model.predict(X_test)

# Valuta le prestazioni del modello di Gradient Boosting
gb_mse = mean_squared_error(y_test, gb_y_pred)
gb_r2 = r2_score(y_test, gb_y_pred)

print("Gradient Boosting - Mean Squared Error:", gb_mse)
print("Gradient Boosting - R-squared:", gb_r2)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, r2_score

##NEURAL
# Carica il dataset
df = pd.read_csv("C:/Users/Utente/Desktop/cakio/dcereijo-player-scores/data/transfers.csv")

# Rimuovi eventuali righe con valori mancanti
df = df.dropna()

# Seleziona le colonne che verranno utilizzate come features (variabili indipendenti)
features =  df[['club_name', 'player_name', 'age', 'position', 'club_involved_name', 'transfer_period', 'league_name', 'season', 'OverallSeasonClub', 'OverallSeasonClub2']]

# Codifica le variabili categoriche
features_encoded = pd.get_dummies(features)

# Seleziona la variabile target (variabile dipendente)
target = df['fee_cleaned']

# Suddividi il dataset in set di addestramento e di test
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Standardizzazione delle features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Crea il modello di rete neurale
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

# Compila il modello
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Addestra il modello
model.fit(X_train_scaled, y_train, epochs=100, batch_size=32, verbose=0)

# Effettua le previsioni sul set di test
y_pred = model.predict(X_test_scaled)

# Valuta le prestazioni del modello
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
