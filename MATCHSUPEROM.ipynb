```python
import pandas as pd

# Leggi i file CSV
match360_df = pd.read_csv(r"C:\Users\Utente\Desktop\cakio\dcereijo-player-scores\data\MATCHfinal4.csv")
teamtry235_df = pd.read_csv(r"C:\Users\Utente\Desktop\cakio\dcereijo-player-scores\data\LEAGUEmean1.csv")

# Unisci i dataframe sulla base delle colonne "home_club_id" e "_id"
merged_df = pd.merge(match360_df, teamtry235_df, left_on='competition_id', right_on='competiton_id')

# Seleziona le colonne desiderate
columns = ['club_home_name', 'club_away_name','name22', 'competition_id', 'round', 'season',
           'home_club_goals', 'away_club_goals', 'away_club_id', 'home_club_id',
           'game_id', 'peso round', 'Overall_home', 'Overall_away', 'mean_overall']


# Crea il dataframe finale senza le colonne indesiderate
final_df = merged_df[columns]
# Cambia il nome della colonna
final_df.rename(columns={'mean_overall': 'Overall_league'}, inplace=True)
# Visualizza il dataframe finale
print(final_df)
```

```python

# Calcola il risultato dell'equazione per ogni riga del dataframe
final_df['Overall_Match_Home'] = ((final_df['Overall_home'] + final_df['Overall_away']) * final_df['Overall_league']) * final_df['peso round']

# Aggiungi il 10% se home_club_goals > away_club_goals
final_df.loc[final_df['home_club_goals'] > final_df['away_club_goals'], 'Overall_Match_Home'] += final_df['Overall_Match_Home'] * 0.1

# Sottrai il 10% se home_club_goals < away_club_goals
final_df.loc[final_df['home_club_goals'] < final_df['away_club_goals'], 'Overall_Match_Home'] -= final_df['Overall_Match_Home'] * 0.1

# Visualizza il dataframe con la nuova colonna
print(final_df)
```

```python
# Calcola il risultato dell'equazione per ogni riga del dataframe
final_df['Overall_Match_Away'] = ((final_df['Overall_home'] + final_df['Overall_away']) * final_df['Overall_league']) * final_df['peso round']

# Aggiungi 0.05 se home_club_goals < away_club_goals
final_df.loc[final_df['home_club_goals'] < final_df['away_club_goals'], 'Overall_Match_Away']  += final_df['Overall_Match_Away'] * 0.1

# Sottrai 0.05 se home_club_goals > away_club_goals
final_df.loc[final_df['home_club_goals'] > final_df['away_club_goals'], 'Overall_Match_Away'] -= final_df['Overall_Match_Away'] * 0.1

# Visualizza il dataframe con la nuova colonna
print(final_df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Crea un oggetto scaler
scaler = MinMaxScaler()

# Reshape della colonna in una matrice 2D
overall_match_values = final_df['Overall_Match_Away'].values.reshape(-1, 1)

# Standardizza i valori nel range da 0 a 1
overall_match_scaled = scaler.fit_transform(overall_match_values)

# Assegna i valori standardizzati alla colonna "Overall_Match" nel dataframe
final_df['Overall_Match_Away'] = overall_match_scaled

# Visualizza il dataframe con la colonna standardizzata
print(final_df)
```

```python
# Crea un oggetto scaler
scaler = MinMaxScaler()

# Reshape della colonna in una matrice 2D
overall_match_values = final_df['Overall_Match_Home'].values.reshape(-1, 1)

# Standardizza i valori nel range da 0 a 1
overall_match_scaled = scaler.fit_transform(overall_match_values)

# Assegna i valori standardizzati alla colonna "Overall_Match" nel dataframe
final_df['Overall_Match_Home'] = overall_match_scaled

# Visualizza il dataframe con la colonna standardizzata
print(final_df)
```

```python
import pandas as pd

# Somma i valori di Overall_Match_Home e Overall_Match_Away raggruppati per anno e squadra
result_df = final_df.groupby(['season', 'name22', 'club_away_name'])['Overall_Match_Home', 'Overall_Match_Away'].sum().reset_index()

# Droppa i risultati duplicati mantenendo solo una singola occorrenza
merged_df = merged_df.drop_duplicates()

# Visualizza il dataframe risultante
print(result_df)
```

```python
result_df['Overall_year_sum'] = result_df['Overall_Match_Home'] + result_df['Overall_Match_Away']
print(result_df)
```

```python
import pandas as pd

# Calcola il valore minimo e massimo per ogni anno
min_values = result_df.groupby('season')['Overall_year_sum'].min()
max_values = result_df.groupby('season')['Overall_year_sum'].max()

# Standardizza i valori da 0 a 1 in base all'anno
result_df['Overall_YEAR'] = result_df.groupby('season')['Overall_year_sum'].apply(lambda x: (x - min_values[x.name]) / (max_values[x.name] - min_values[x.name]))

# Visualizza il dataframe risultante
print(result_df)
```

```python
result_df = result_df[['season', 'name22', 'Overall_YEAR']]
print(result_df)
```
