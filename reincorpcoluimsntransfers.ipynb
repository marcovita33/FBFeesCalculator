```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

# Carica il file CSV in un DataFrame
df = pd.read_csv('transfers.csv')

# Crea un nuovo DataFrame con le colonne desiderate e valori unici
new_df = df[['club_name', 'league_name']].drop_duplicates()

# Funzione per trovare il "miglior match"
def find_best_match(row):
    club_name = row['club_involved_name']
    best_match = process.extractOne(club_name, new_df['club_name'])
    if best_match[1] >= 95:
        return new_df.loc[best_match[2], 'league_name']
    else:
        return None

# Aggiungi la colonna "league_destination" al DataFrame originale
df['league_destination'] = df.apply(find_best_match, axis=1)

# Stampa il DataFrame con la colonna "league_destination" aggiunta
print(df)
```
