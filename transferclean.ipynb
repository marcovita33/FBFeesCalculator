```python
import pandas as pd

file_paths = [
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\primera-division.csv",
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\premier-liga.csv",
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\premier-league.csv",
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\ligue-1.csv",
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\liga-nos.csv",
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\eredivisie.csv",
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\1-bundesliga.csv",
    "C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\serie-a.csv"
]

dfs = []

# Carica i dataframe dai file CSV e rimuovi le righe precedenti all'anno 2012
for file_path in file_paths:
    df = pd.read_csv(file_path)
    df = df[df['year'] >= 2012]
    dfs.append(df)

# Unisci tutti i dataframe in uno solo
merged_df = pd.concat(dfs)

# Visualizza il dataframe risultante
print(merged_df)
```

```python
import pandas as pd
from fuzzywuzzy import process
from fuzzywuzzy import fuzz

# Carica il primo file CSV
df1 = pd.read_csv("C:\\Users\\Utente\\Desktop\\cakio\\dcereijo-player-scores\\data\\overallxseason.csv")

# Carica il secondo file CSV
df2 = merged_df

```

```python


def find_best_match(row):
    name = row['club_name']
    year = row['year']
    df1_filtered = df1[df1['season'] == year]  # Filtra df1 per l'anno corrispondente
    matches = process.extractOne(name, df1_filtered['name22'], scorer=fuzz.token_sort_ratio)
    if matches:
        return matches[1], df1_filtered.loc[matches[2], 'name22'], df1_filtered.loc[matches[2], 'Overall_YEAR']

df2['best_match'] = df2.apply(find_best_match, axis=1)

# Divide i risultati del miglior match in colonne separate
df2[['match_score', 'best_match_name', 'OverallSeasonClub']] = pd.DataFrame(df2['best_match'].tolist(), index=df2.index)

# Rimuovi la colonna temporanea 'best_match'
df2 = df2.drop('best_match', axis=1)

# Assegna 0 a OverallSeasonClub per le righe con match_score < 80
df2.loc[df2['match_score'] < 80, 'OverallSeasonClub'] = 0

print(df2)
```

```python
from fuzzywuzzy import process
from fuzzywuzzy import fuzz

def find_best_match(row):
    name = row['club_involved_name']
    year = row['year']
    df1_filtered = df1[df1['season'] == year]  # Filtra df1 per l'anno corrispondente
    matches = process.extractOne(name, df1_filtered['name22'], scorer=fuzz.token_sort_ratio)
    if matches:
        return matches[1], df1_filtered.loc[matches[2], 'name22'], df1_filtered.loc[matches[2], 'Overall_YEAR']

df2['best_match'] = df2.apply(find_best_match, axis=1)

# Divide i risultati del miglior match in colonne separate
df2[['match_score', 'best_match_name', 'OverallSeasonClub2']] = pd.DataFrame(df2['best_match'].tolist(), index=df2.index)

# Assegna 0 a OverallSeasonClub per le righe con match_score < 80
df2.loc[df2['match_score'] < 80, 'OverallSeasonClub2'] = 0

# Rimuovi la colonna temporanea 'best_match'
df2 = df2.drop('best_match', axis=1)
print(df2)
```

```python
# Converti i valori della colonna 'fee' in stringhe
df2['fee'] = df2['fee'].astype(str)

# Rimuovi le righe che contengono le parole "loan" o "free" nella colonna "fee"
df2 = df2.loc[~df2['fee'].str.contains('loan|free', case=False)]

# Rimuovi le righe in cui il valore nella colonna "transfer_movement" è "out"
df2 = df2.loc[df2['transfer_movement'] != 'out']
```

```python
# Rimuovi le righe che presentano valori NaN nella colonna 'fee_cleaned'
df2 = df2.dropna(subset=['fee_cleaned'])


```

```python
# Rimuovi le colonne 'fee', 'transfer_movement', 'year', 'country' e 'match_score'
df2 = df2.drop(['fee', 'transfer_movement', 'year', 'country','best_match_name',  'match_score'], axis=1)

print(df2)

```

```python
# Rimuovi le righe con valori NaN
df2.dropna(inplace=True)

# Stampa il DataFrame risultante
print(df2)
```
