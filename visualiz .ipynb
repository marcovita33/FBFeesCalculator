```python
import pandas as pd

# Leggi il file CSV
file_path = r"C:\Users\Utente\Desktop\cakio\dcereijo-player-scores\data\MATCHGIGACHAD.csv"
df = pd.read_csv(file_path)

# Filtra il dataframe per le squadre desiderate
squadre = ['Real Madrid', 'Bayern Munich', 'Manchester City', 'Liverpool', 'Barcelona',
           'Paris Saint-Germain', 'Ajax', 'Borussia Dortmund', 'Manchester United', 'Arsenal',
           'Juventus', 'Atletico Madrid', 'Chelsea', 'Sevilla FC', 'Napoli',
           'Tottenham Hotspur', 'Internazionale', 'Villarreal', 'AS Roma']
filtered_df = df[df['name22'].isin(squadre)]

# Trova il valore massimo di "Overall_home" per ciascuna squadra
max_overall_home = filtered_df.groupby('name22')['Overall_away'].max()

# Stampa i risultati
for squadra, overall_home in max_overall_home.items():
    print(f"{squadra}: {overall_home}")
```

```python
import pandas as pd

# Leggi il file CSV
file_path = r"C:\Users\Utente\Desktop\cakio\dcereijo-player-scores\data\overallxseason.csv"
df = pd.read_csv(file_path)

# Filtra il dataframe per le squadre desiderate
squadre = ['Real Madrid', 'Bayern Munich', 'Manchester City', 'Liverpool', 'Barcelona',
           'Paris Saint-Germain', 'Ajax', 'Borussia Dortmund', 'Manchester United', 'Arsenal',
           'Juventus', 'Atletico Madrid', 'Chelsea', 'Sevilla FC', 'Napoli',
           'Tottenham Hotspur', 'Internazionale', 'Villarreal', 'AS Roma']
filtered_df = df[df['name22'].isin(squadre)]

# Trova la stagione con il valore massimo di "Overall_YEAR" per ciascuna squadra
max_overall_season = filtered_df.groupby('name22')['Overall_YEAR'].idxmax()
seasons = filtered_df.loc[max_overall_season, ['name22', 'Overall_YEAR', 'season']]

# Stampa i risultati
for index, row in seasons.iterrows():
    squadra = row['name22']
    overall_year = row['Overall_YEAR']
    stagione = row['season']
    print(f"{squadra}: Overall_YEAR = {overall_year} (Stagione: {stagione})")
```

```python
import pandas as pd

# Leggi il file CSV
df = pd.read_csv(r"C:\Users\Utente\Desktop\cakio\dcereijo-player-scores\data\transfers3.csv")

# Seleziona le colonne da mantenere
columns_to_keep = ['club_name', 'player_name', 'age', 'club_involved_name', 'fee_cleaned', 'season']

# Crea un nuovo dataframe con le colonne selezionate
df_filtered = df[columns_to_keep]

# Ordina il dataframe in base al valore della colonna "fee_cleaned" in ordine decrescente
df_sorted = df_filtered.sort_values(by='fee_cleaned', ascending=False)

# Visualizza i primi 20 risultati
top_20 = df_sorted.head(25)
print(top_20)
```

```python
import matplotlib.pyplot as plt

# Crea il grafico a barre orizzontale
plt.figure(figsize=(8, 10))
bars = plt.barh(top_20['player_name'], top_20['fee_cleaned'])
plt.xlabel('Fee Cleaned')
plt.ylabel('Player')
plt.title('Top 20 Players by Fee Cleaned')

# Aggiungi il nome del club come testo sulla barra corrispondente
for i, bar in enumerate(bars):
    plt.text(bar.get_width() - 10, bar.get_y() + bar.get_height() / 2, top_20['club_name'].iloc[i],
             color='white', va='center', ha='right')

# Mostra il grafico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crea il grafico a barre orizzontale
plt.figure(figsize=(10, 12))
bars = plt.barh(range(len(top_20)), top_20['fee_cleaned'])
plt.xlabel('Fee Cleaned')
plt.ylabel('Player')
plt.title('Top 20 Players by Fee Cleaned')

# Etichetta le barre con il nome del club, il nome del giocatore e la stagione
for i, bar in enumerate(bars):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f"{top_20['club_name'].iloc[i]} ({top_20['season'].iloc[i]})",
             color='white', va='center', ha='right')  # Modificato da 'right' a 'left'

# Aggiungi i nomi dei giocatori esternamente sull'asse Y
plt.yticks(range(len(top_20)), top_20['player_name'])

# Sposta i nomi dei club a sinistra
plt.subplots_adjust(left=0.3)

# Mostra il grafico
plt.tight_layout()
plt.show()
```
